```python
import matplotlib.pyplot as plt
import random as rnd
import numpy as np
import math
import sympy
from decimal import Decimal


def alpha_values(array):
  while(len(array)<5):
    y = rnd.randrange(1,99)
    if y not in array:
      array.append(y)

# Poisson(alpha) Distribution is a limiting case of Binomial(n,p) Distribution 
# We can establish the relation Alpha = n*p majorly for large values of n and small values of p that is np
#

def binomial_values(alpha):
  p = 0.1
  n = int(alpha//p)
  #n = alpha + 1
  #p = round(alpha/n,2)
  bi = [n,p]
  return bi

def graphs(n,p,alpha):
  result_array_bi = []
  result_array_poi = []
  for i in range(n+1):

    k = (math.factorial(n)/(math.factorial(i)*math.factorial(n-i)))*((p)**i)*((1-p)**(n-i))
    #num = math.factorial(n) * round(((Decimal(p))**i * (Decimal(1)-Decimal(p))**(n-i)),10)
    #denom = math.factorial(i) * math.factorial(n-i)
    #k = Decimal(num) / Decimal(denom)
    result_array_bi.append(k)

  i = 0
  if (alpha<75):
    while(i<100):
      x = ((alpha**i)*(math.e**(-1*alpha)))/(math.factorial(i))
      result_array_poi.append(x)
      i=i+1
  else:
    i = 50
    while(i<150):
      x = ((alpha**i)*(math.e**(-1*alpha)))/(math.factorial(i))
      result_array_poi.append(x)
      i=i+1
  
  y = np.array([i for i in range(n+1)])
  plt.bar(y,result_array_bi,width = 0.75)
  plt.xlabel('Value in Range Space', fontsize=12, color='blue', fontstyle='italic')
  plt.ylabel('Probability', fontsize=12, color='blue', fontstyle='italic')
  plt.title('PMF of Binomial ('+str(n)+','+str(p)+') RV', fontsize=14)
  plt.show()

  if (alpha<75):
    y = np.array([i for i in range(100)])
  else:
    y = np.array([i for i in range(50,150)])
  plt.bar(y,result_array_poi,width=0.25)
  plt.xlabel('Value in Range Space', fontsize=12, color='blue', fontstyle='italic')
  plt.ylabel('Probability', fontsize=12, color='blue', fontstyle='italic')
  plt.title('PMF of Poisson ('+str(alpha)+') RV', fontsize=14)
  plt.show()

  #print(result_array_bi)
  #print(result_array_poi)

array = []
alpha_values(array)
binomial_val_list = []
temp_list = []

for i in range(4):
  temp_list = binomial_values(array[i])
  binomial_val_list.append(temp_list)
  n = temp_list[0]
  p = temp_list[1]
  graphs(n,p,array[i])

print(array)
print(binomial_val_list)
```
